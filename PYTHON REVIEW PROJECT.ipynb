```python
#Import the libraries 

import pandas as pd
import numpy as np
import os
os.chdir('/Users/rebaviles21/QBIO_490_rebecca/analysis_data')
```

```python
import cptac
cptac.download(dataset="Ccrcc")
ccrcc = cptac.Ccrcc()
```

```python
clinical_data = ccrcc.get_clinical()
clinical_data
```

```python
protein_data = ccrcc.get_proteomics()
protein_data.columns = protein_data.columns.get_level_values(0) 
protein_data
```

```python
rna_data = ccrcc.get_transcriptomics()
rna_data
```

```python
unite_data = pd.merge(protein_data, clinical_data, on='Patient_ID')
unite_data
```

```python
#Preprocessing and cleaning the data
clinical_data = clinical_data[clinical_data['tumor_stage_pathological'].notna()]
ccrcc_index = clinical_data.index
protein_data = protein_data.loc[ccrcc_index]

#clean up the stages CRCC
stage1_mask = clinical_data['tumor_stage_pathological'] == "Stage I"

stage3_mask = clinical_data['tumor_stage_pathological'] == "Stage III"

#clean up for proteins 
protein_stage1 = protein_data.loc[stage1_mask]

protein_stage3 = protein_data.loc[stage3_mask]


#collect the mean values to prepare the between cleaned data
first_protein = protein_stage1.mean()

third_protein = protein_stage3.mean()


#find the corresponding and most relevant data 

differential_expression = np.abs(first_protein - third_protein)

#column created to find the proteins
protein_sorting = differential_expression.sort_values(ascending = False) 
cleaned_protein_sorting = protein_sorting.dropna()

#collecting using RNA and clinical 
five_proteins = cleaned_protein_sorting.head(5)

print(five_proteins)

```

```python
# 2) Create a new dataframe of your selected features, where the rows are the patients
#and the columns are the expression values of genes you selected in step 1 (X data).

import numpy as np
import pandas as pd

rna_data = rna_data.loc[ccrcc_index]
log_rna = np.log2(rna_data)

#set up cleaned masks to find the appropriate genes

genes_stage1 = log_rna.loc[stage1_mask]
genes_stage3 = log_rna.loc[stage3_mask]

genes_first = genes_stage1.mean()
genes_third = genes_stage3.mean()
#differential expression to find highest mean value of genes 

gene_differential_expression = np.abs(genes_first - genes_third)

gene_sorting = gene_differential_expression.sort_values(ascending=False)
cleaned_gene_sorting = gene_sorting.dropna()

five_genes = cleaned_gene_sorting.head(5)

print(five_genes)
```

```python
selected_genes = ['LRRC43', 'SYNPO2L', 'C1orf145', 'MAST1', 'SYTL5']
chosen_selected = log_rna[selected_genes]
print(chosen_selected)
```

```python
#3) create a separate list of the patients' cancer stage , (tumor_stage_pathological)
cancer_stage = clinical_data['tumor_stage_pathological']

#inconsistencies in step 6 with mistmatched due to not clean data
cancer_mask = ['C3L-00097', 'C3L-00448', 'C3L-00796', 'C3L-01607', 'C3N-00313', 'C3N-01178']
new_mask = ~cancer_stage.index.isin(cancer_mask)
newCancer_stage= cancer_stage[new_mask]

print(newCancer_stage)
```

```python
#4) scale and encode your features and target
#import graphing libraries and tools 

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

chosen_selected.replace([np.inf, -np.inf], np.nan, inplace = True)
chosen_selected = chosen_selected.dropna()

#scaling 
scaler = StandardScaler()
scaled_feature = scaler. fit_transform(chosen_selected)

#encoding 
label = LabelEncoder()
label_encoder_features = label.fit_transform(cancer_stage)

print(chosen_selected)
print(newCancer_stage)
```

```python
#5) create a train split of the X and Y data with train_size = 0.7

X_train, X_test, y_train, y_test = train_test_split(chosen_selected, newCancer_stage, train_size=0.7, random_state=42)
```

```python
#6) Test the accuracy of the four models and find the average accuracy between them all

#import the data needed for the models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

kneighbors_class = KNeighborsClassifier()
decisiontree_class = DecisionTreeClassifier()
mlp_class = MLPClassifier()
gaussianNB_class = GaussianNB()

amt_of_runs = 10
accuracy = []
totalScore = 0

for _ in range(amt_of_runs):
    X_train, X_test, y_train, y_test = train_test_split(chosen_selected, newCancer_stage, train_size=0.7, random_state=42)

    
    
#KNeighbors Model 

kneighbors_class.fit(X_train, y_train)
knn_predictor = kneighbors_class.predict(X_test)
knn_accurate = accuracy_score(y_test, knn_predictor)
knNeighbors = kneighbors_class.score(np.array(X_test), y_test)


#DecisionTree Model

decisiontree_class.fit(X_train, y_train)
decisiontree_predictor = decisiontree_class.predict(X_test)
decisiontree_accurate = accuracy_score(y_test, decisiontree_predictor)
decisionTree = decisiontree_class.score(np.array(X_test), y_test)

#MLPClassifier Model
mlp_class.fit(X_train, y_train)
mlp_predictor = mlp_class.predict(X_test)
mlp_accurate = accuracy_score(y_test, mlp_predictor)
mlp = mlp_class.score(np.array(X_test), y_test)

#GaussianNB Model

gaussianNB_class.fit(X_train, y_train)
gaussianNB_predictor = gaussianNB_class.predict(X_test)
gaussianNB_accurate = accuracy_score(y_test, gaussianNB_predictor)
gaussianNB = gaussianNB_class.score(np.array(X_test), y_test)

#set up the runs and store the accuracy scores
print(knNeighbors)
print(decisionTree)
print(mlp)
print(gaussianNB)

#In this case, the Gaussian NB classifier is the best one! This is then followed by MLP classifier, KNNeigbors Classifier, and then the least acccurate was the Decision Tree
```
